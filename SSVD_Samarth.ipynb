```python
import pdfplumber

all_sentences = []
with pdfplumber.open(r'DCISC-24th-Annual-Report.pdf') as pdf:
    page_numbers = len(pdf.pages)
    for num in range(page_numbers):
        page = pdf.pages[num]
        page_text = page.extract_text()
        page_text = page_text.replace('\n'," ")
        page_sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', page_text)
        all_sentences.append(page_sentences)
```

```python
print(len(all_sentences))
```

```python
clean_articles=[]
cl2=[]
cl3=[]
for i in range(len(all_sentences)):
  cl2.append([re.sub("[^0-9a-zA-Z]"," ",x) for x in all_sentences[i]])
for i in range(len(cl2)):
  cl3.append([' '.join([w for w in x.split() if len(w)>1]) for x in cl2[i]])
for i in range(len(cl3)):
  clean_articles.append([x.lower() for x in cl3[i]])
```

```python
stop_words = nlp.Defaults.stop_words

# assume that 'df' is a list of lists containing strings
for i in range(len(clean_articles)):
    # join the list into a single string
    text = ' '.join(clean_articles[i])
    
    # remove single character terms
    terms_list = [word for word in text.split() if len(word) > 2]
    
    # lemmatization
    lemma_list = [token.lemma_ for token in nlp(' '.join(terms_list))]
    
    # remove stopwords and -PRON- tags
    clean_list = [re.sub('-PRON-', '', word) for word in lemma_list if word not in stop_words]
    
    # join the cleaned tokens back into a string
    clean_text = ' '.join(clean_list)
    
    # assign the cleaned text back to 'df'
    df[i] = clean_text
```

```python
stop_words = nlp.Defaults.stop_words

# create a nested list to store each page as a list of terms
nl = []

# assume that 'df' is a list of lists containing strings
for i in range(len(clean_articles)):
    # join the list into a single string
    text = ' '.join(clean_articles[i])
    
    # remove single character terms
    terms_list = [word for word in text.split() if len(word) > 2]
    
    # lemmatization
    lemma_list = [token.lemma_ for token in nlp(' '.join(terms_list))]
    
    # remove stopwords and -PRON- tags
    clean_list = [re.sub('-PRON-', '', word) for word in lemma_list if word not in stop_words]
    
    # join the cleaned tokens back into a string
    clean_text = ','.join(clean_list)
    
    # append the cleaned text to the nested list
    nl.append(clean_text.split(','))

# print the nested list
print(nl)
```

```python
def get_topics(num_topics, num_top_words, svd_model, vectorizer, seed_words=None):
    """
    Obtain the top words for each topic using the SVD model and vectorizer.
    If seed words are provided, use them to initialize the SVD model.
    """
    if seed_words is not None:
        # Use the seed words to initialize the SVD model
        seed_matrix = vectorizer.transform(seed_words)
        svd_model.fit(seed_matrix)
    
    # Get the singular vectors for the matrix using the SVD model
    svd_matrix = svd_model.components_
    
    # Get the top words for each topic by sorting the singular vectors
    topic_words = []
    for i in range(num_topics):
        singular_vector = svd_matrix[i, :]
        top_word_indices = singular_vector.argsort()[::-1][:num_top_words]
        top_words = [vectorizer.get_feature_names_out()[index] for index in top_word_indices]
        topic_words.append(top_words)

        print(top_words)

    return topic_words,svd_model
```
